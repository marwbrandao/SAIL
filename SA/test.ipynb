```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create a figure and axis
fig, ax = plt.subplots()

# Plot the data
ax.plot(x, y)

# Show the plot
plt.show()
```

```python
# Data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Create a figure and axis
fig, ax = plt.subplots()

# Plot the data
ax.plot(x, y1, label='sin')
ax.plot(x, y2, label='cos')

# Add a legend
ax.legend()

# Show the plot
plt.show()
```

```python
# Data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Create a figure and axes
fig, ax = plt.subplots(2)

# Plot the data
ax[0].plot(x, y1)
ax[1].plot(x, y2)

# Show the plot
plt.show()
```
